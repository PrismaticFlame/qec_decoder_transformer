```python
# getting this info from: https://nordiquest.net/application-library/training-material/qas2024/notebooks/surface_code_threshold.html
import stim

circuit = stim.Circuit.generated("surface_code:rotated_memory_x", rounds=2, distance=3)
# This will create a distance 3 code, and 3 rounds of syndrome measurement. This is a predefined quantum circuit, not one we have to make.
```

This block below will generate an image of a repetition code *of* a surface code. For a distance 3 code, we should expect to see 9 data qubits and 8 ancilla qubits, for a total of 15 qubits. The code also generates a number of idle wires which correspond to unused qubits.

```python
circuit.diagram("timeline-svg")
```

However, this isn't a great way of viewing things is utiliing a **timeslice** diagram of the circuit. A timeslice diagram depicts the parts of the circuit executed between two `TICK` operations.

```python
circuit.diagram("timeslice-svg")
```
